### Problem 2 – Constraint Logic

Using the same matrix D:
1. Normalize each row by subtracting its row mean and dividing by its row standard deviation.
2. Remove columns where more than 40% of values are zero.
3. From the remaining matrix, extract the submatrix formed by:
    - rows with positive mean
    - columns whose maximum is above the overall mean

This forces dimension-based conditional logic.

---

Doing 

```
D -= row_mean
D = row_std/D
```
in two steps because 

$$\frac{(a-b)}{c} = \left(\frac{a}{c}-\frac{b}{c}\right) \: \neq \: \frac{ca - b}{c} = a - \frac{b}{c}$$

````python
#1
row_mean = np.mean(D,axis=1)[:,None]
row_std = np.mean(D,axis=1)[:,None]

D -= row_mean
D = row_std/D
D
````

````python
#2
percent40ValuesCount = D.shape[0] * 0.4
cond = D == 0
mask = np.sum(cond,axis=0) != 4
D[:,mask]
````

````python
#3
# We need to make two masks
# Row mask and column mask 
# D[np.mean(D,axis=1) > 0,np.max(D,axis=0) > np.mean(D)] # This didn't work cause numpy tryna zip two arrays but they are not of same length
D[np.mean(D,axis=1) > 0][:,np.max(D,axis=0) > np.mean(D)] # apply masks one by one
````

````python
subjects = [('roll no',np.int64),
            ('ala',np.int32),
            ('agt',np.int32),
            ('ca',np.int32),
            ('E waste',np.int32),
            ('green chem',np.int32),
            ('rm',np.int32)]
scores0 = np.random.randint(0,100,size=(50,6))
````

````python
zipped = {k:v  for k,v in zip(np.arange(1,51),scores0)}
a = ((k,*list(v)) for k,v in  zipped.items())
````

````python
scores = np.array(list(a),dtype=subjects)
````

````python
# I keep these cells here because it reminds me of what I,ve tried experimenting with
# Define a structured data type with field names and types
# 'name' -> string of max length 20
# 'age'  -> integer
# 'marks' -> float
student_dtype = np.dtype([
    ('name', 'U20'),   # Unicode string of length 20
    ('age', 'i4'),     # 32-bit integer
    ('marks', 'f4')    # 32-bit float
])

# Create a structured array with initial data
students = np.array([
    ['Alice', 21, 88.5],
    ['Bob', 19, 92.0],
    ['Charlie', 22, 79.5]
], dtype=student_dtype)

# Access entire columns by field name
print("Names:", students['name'])
print("Ages:", students['age'])
print("Marks:", students['marks'])

# Access a single row
print("First student record:", students[0])

# Modify a field
students[1]['marks'] = 95.0
print("Updated Bob's marks:", students[1])

# Filtering based on a condition
high_scorers = students[students['marks'] > 85]
print("High scorers:\n", high_scorers)

# Adding a new record (requires creating a new array)
new_student = np.array([('David', 20, 81.0)], dtype=student_dtype)
students = np.concatenate((students, new_student))
print("After adding David:\n", students)
````
