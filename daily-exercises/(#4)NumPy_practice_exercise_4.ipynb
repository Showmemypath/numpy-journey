### Problem 1 – Global Ranking + Structure

1. Create a matrix D of shape (10, 12), values between -100 and 200.
2. Replace all negative values with the median of the entire matrix.
3. Keep only the top 20% largest values per column, set the rest to 0.
4. Reorder rows based on their variance (highest variance first).
5. Create a new matrix containing only rows whose maximum value is above the 90th percentile of the entire matrix.

````python
import numpy as np
````

````python
#1
def resetD():
    np.random.seed(42)
    global D
    D = np.random.randint(-100,200,size=(10,12))
    D = D.astype(np.float64)
resetD()
````

````python
#2
print(np.where(D<0,np.median(D),D))
print("Median:",np.median(D))
````

````python
#3
col_length = D.shape[0]
percent20 = np.int64(col_length * (20/100))

a = np.arange(0,12)
multiplier = np.ones((8,12))

c = np.int64(np.multiply(a,multiplier))
b = np.argsort(D,axis=0)[:8]

D[b,c]=0
print(D)

# I did this by advanced indexing , getting the indices of values which are not the 2 two largest values of that column and setting it to 0
````

### Problem 2 – Constraint Logic

Using the same matrix D:
1. Normalize each row by subtracting its row mean and dividing by its row standard deviation.
2. Remove columns where more than 40% of values are zero.
3. From the remaining matrix, extract the submatrix formed by:
4. rows with positive mean
5. columns whose maximum is above the overall mean
6. This forces dimension-based conditional logic.
---

Doing 

```
D -= row_mean
D = row_std/D
```
in two steps because 

$$\frac{(a-b)}{c} = \left(\frac{a}{c}-\frac{b}{c}\right) \: \neq \: \frac{ca - b}{c} = a - \frac{b}{c}$$

````python
#1
row_mean = np.mean(D,axis=1)[:,None]
row_std = np.mean(D,axis=1)[:,None]

D -= row_mean
D = row_std/D
D
````

````python
#2
percent40ValuesCount = D.shape[0] * 0.4
cond = D == 0
mask = np.sum(cond,axis=0) != 4
D[:,mask]
````

````python
#3
# We need to make two masks
# Row mask and column mask 
# D[np.mean(D,axis=1) > 0,np.max(D,axis=0) > np.mean(D)] # This didn't work cause numpy tryna zip two arrays but they are not of same length
D[np.mean(D,axis=1) > 0][:,np.max(D,axis=0) > np.mean(D)] # apply masks one by one
````
