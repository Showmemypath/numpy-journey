### Question 11 (sorting)
Given X = np.array([1,1,2,2]) and Y = np.array([3,1,4,2]), explain why the order of keys matters in lexsort.

```python
X = np.array([1,1,2,2])
Y = np.array([3,1,4,2])
print("Order of keys matter because the last key is the primary sorting key which is getting sorting")
print("If there are equal values then the values in the previous key ( that comes before it ) are compared to sort the keys of the primary key")
print("and lexsort returns the index")
print("See the differnence ")
print("np.lexsort((X,Y))",f"index :{np.lexsort((X,Y))}",f"sorted : {Y[np.lexsort((X,Y))]}")
print("np.lexsort((Y,X))",f"index :{np.lexsort((Y,X))}",f"sorted : {X[np.lexsort((Y,X))]}")
```
