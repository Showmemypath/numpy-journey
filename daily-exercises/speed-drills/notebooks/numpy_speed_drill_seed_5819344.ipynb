### Question 2 (broadcasting)
Given A = np.arange(6).reshape(2,3) and B = np.array([[1],[2]]), determine the result of A + B.

```python
A = np.arange(6).reshape(2,3) # 2 rows 3 cols
B = np.array([[1],[2]]) # 2 rows 1 col 
# B will stretch and becomes [[1,1,1],[2,2,3]]
# Now element wise addition will be done
A + B
```

### Question 10 (broadcasting)
Given A = np.arange(12).reshape(3,4), explain why broadcasting does not copy data.

```python
# Broadcasting does not copy data because Numpy implements it by modifying array meta deta ( shape + strides ), not by allocating new memory
# Multiple elements share the same memory location because of this broadcasted arrays are normally read only views
```
